```python
import pandas as pd
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, roc_curve, auc

from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import cross_val_score

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# load dataset
data=pd.read_csv('drive/MyDrive/Research/gene data.csv')
label=pd.read_csv('drive/MyDrive/Research/labels.csv')
```

```python
class_labels=label["Class"]
class_labels
```

```python
class_labels.value_counts().plot(kind='bar',color=['firebrick', 'green', 'blue', 'black', 'red'], figsize=(4,3))
```

```python
label = label.replace({'PRAD': 0})
label = label.replace({'LUAD': 1})
label = label.replace({'BRCA': 2})
label = label.replace({'KIRC': 3})
label = label.replace({'COAD': 4})
```

```python
label.head()
```

```python
X=data.iloc[:,1:]
y=label.iloc[:,1:]
```

```python
y
```

```python
dataset=pd.concat([X, class_labels], axis=1)
```

```python
dataset = dataset.set_index('Class').sort_index()
```

```python
dataset.head()
```

```python
dt=dataset
```

```python
missing=[]
for i in range(len(dataset.columns)):
    if((dt['gene_'+str(i)]==0).all()):
        missing.append(i)
```

```python
print(len(missing))
```

```python
for i in range(len(missing)):
    print(missing[i],end=' ')
```

```python
for i in range(len(missing)):
    X=X.drop(['gene_'+str(missing[i])], axis = 1)
```

```python
X=pd.DataFrame(X).to_numpy()
y=pd.DataFrame(y).to_numpy()
```

```python
target=y
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(target)
onehot_encoder = OneHotEncoder(sparse_output=False, categories='auto')
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
target_encoded = onehot_encoder.fit_transform(integer_encoded)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,target_encoded,test_size=0.20,random_state=42,shuffle=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
y_test = np.argmax(y_test, axis=1)
y_train = np.argmax(y_train, axis=1)
```

## Random Forest Classification

```python
# Create Random Forest classifier object
rf = RandomForestClassifier(n_estimators=100,max_depth=5,min_samples_split=4,min_samples_leaf=2,max_features='sqrt',random_state=42)
# Train Random Forest Classifier
rf = rf.fit(X_train, y_train)
# Predict the response for test dataset
y_pred = rf.predict(X_test)
# Model Accuracy, how often is the classifier correct?
print("Accuracy on RF: %.2f" % (accuracy_score(y_test, y_pred)*100),"%")
```

```python
print(classification_report(y_test, y_pred, digits=4))
```

```python
scross = cross_val_score(rf,X,y, cv = 5)
```

```python
print(scross)
print(scross.mean())
```

```python
cm_rf = confusion_matrix(y_test, y_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=class_labels.unique())
disp.plot()
plt.title("Confusion Matrix - Random Forest")
plt.show()
```

```python
# Generate ROC curve for Random Forest
fpr, tpr, thresholds = roc_curve(y_test, y_pred, pos_label=2)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - Random Forest')
plt.legend(loc="lower right")
plt.show()
```

## SVM

```python
svm = SVC(C= 1.2, kernel='rbf', gamma='scale', probability=True)
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
```

```python
print(classification_report(y_test, y_pred, digits=4))
```

```python
scross = cross_val_score(svm,X,y, cv = 5)
```

```python
print(scross)
print(scross.mean())
```

```python
cm_svm = confusion_matrix(y_test, y_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm_svm, display_labels=np.unique(y_test))
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - 5-Fold Cross-Validation")
plt.show()
```

##Ensemble (RF, SVM, GB)

```python
ensemble = VotingClassifier(estimators=[
    ('SVM', svm), ('RF', rf), ('GB', gb)],
    voting='soft')
```

```python
ensemble.fit(X_train, y_train)
```

```python
y_pred = ensemble.predict(X_test)
```
